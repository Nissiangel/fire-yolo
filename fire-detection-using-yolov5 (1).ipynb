# Import Packages

```python
import torch
from matplotlib import pyplot as plt
import numpy as np
import cv2
```

# Predict with predefined weights

```python
model = torch.hub.load('ultralytics/yolov5','yolov5s')
```

```python
img = "https://ultralytics.com/images/zidane.jpg"
```

```python
results = model(img)
results.print()
```

```python
%matplotlib inline
plt.imshow(np.squeeze(results.render()))
plt.show()
```
